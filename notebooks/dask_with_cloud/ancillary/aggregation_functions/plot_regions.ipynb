Create a map with analysis regions outlined

```python
import numpy as np
import xarray as xr
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import cartopy

%matplotlib inline
```

```python
# Sub-tropical western Atlantic region:
latbnds1 = (20, 30)
lonbnds1 = (-82, -72)

# Mid-latitude eastern Pacific region:
latbnds2 = (45, 55)
lonbnds2 = (-137, -127)
```

```python
fig = plt.figure(figsize=(10, 4))
ax = plt.axes(projection=cartopy.crs.PlateCarree(central_longitude=0))


# Boxes for regions:
dlat1 = latbnds1[1] - latbnds1[0]
dlon1 = lonbnds1[1] - lonbnds1[0]
dlat2 = latbnds2[1] - latbnds2[0]
dlon2 = lonbnds2[1] - lonbnds2[0]
region1 = patches.Rectangle(
    (lonbnds1[0], latbnds1[0]), dlon1, dlat1, 
    linewidth=1.5, edgecolor='r', facecolor='none'
    )
ax.add_patch(region1)
region2 = patches.Rectangle(
    (lonbnds2[0], latbnds2[0]), dlon2, dlat2, 
    linewidth=1.5, edgecolor='r', facecolor='none'
    )
ax.add_patch(region2)


# Shading for oceans/continents and outlines for coastlines
ax.add_feature(cartopy.feature.OCEAN, zorder=0)
ax.add_feature(cartopy.feature.LAND, zorder=0, edgecolor='black', linewidth=0.5)


# Plotting aesthetics:
ax.set_xlim(-180, 0)
ax.set_ylim(-30, 80)
xticks = np.arange(-180, 1, 30)
ax.set_xticks(xticks, crs=cartopy.crs.PlateCarree())
ax.set_xticklabels(str(t)+r"$^\circ$" for t in xticks)
yticks = np.arange(-30, 81, 30)
ax.set_yticks(yticks, crs=cartopy.crs.PlateCarree())
ax.set_yticklabels(str(t)+r"$^\circ$" for t in yticks)
```

```python
fig.savefig("regions.jpg")
```
