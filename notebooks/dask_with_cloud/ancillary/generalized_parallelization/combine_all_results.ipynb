Combine all results from downscaling computations on various machines/configurations into a single .csv file. Save the .csv file to the current directory.

```python
import os
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
dir_results = "./results/"
filepaths = [dir_results+f for f in os.listdir(dir_results) if f.startswith("computation_results_")]
filepaths
```

```python
results = pd.concat((pd.read_csv(f) for f in filepaths), ignore_index=True)
results.sort_values(by=['ec2_type', 'n_workers', 'threads_per_worker', 'n_files'], inplace=True)
results
```

```python
results.to_csv('dscale_computation_results_all.csv', index=False)
```
